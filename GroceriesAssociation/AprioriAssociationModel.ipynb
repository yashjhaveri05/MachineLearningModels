```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('groceries.csv', header = None)
dataset = dataset.iloc[1:, 1:]
print("Total Rows: {}\nTotal Columns: {}".format(dataset.shape[0],dataset.shape[1]))
dataset.head()
```

```python
groceryBasket = []
for i in range(0, 9835):
  groceryBasket.append([str(dataset.values[i,j]) for j in range(0, 32)])
```

```python
from apyori import apriori
rules = apriori(transactions = groceryBasket, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

```python
results = list(rules)
```

```python
results
```

```python
def inspect(results):
    lhs         = [tuple(result[2][0][0])[0] for result in results]
    rhs         = [tuple(result[2][0][1])[0] for result in results]
    supports    = [result[1] for result in results]
    confidences = [result[2][0][2] for result in results]
    lifts       = [result[2][0][3] for result in results]
    return list(zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ['Left Hand Side', 'Right Hand Side', 'Support', 'Confidence', 'Lift'])
```
